```python
# 20.

def swap_strings(s1, s2):
    new_s1 = s2[:2] + s1[2:]
    new_s2 = s1[:2] + s2[2:]
    return new_s1 + " " + new_s2

print(swap_strings("abc", "xyz"))
```

```python
# 21

def add_suffix(s):
    if len(s) < 3:
        return s
    elif s.endswith("ing"):
        return s + "ly"
    else:
        return s + "ing"
    
print(add_suffix("play"))
print(add_suffix("playing")) 
```

```python
# 22 .

def rev_multi_of(s):
    if len(s) % 4 == 0:
        return s[::-1]
    return s
print(rev_multi_of("abcd"))
```

```python
# 23.

def first_last_two(w):
    if len(w) < 2:
        return""
    return w[:2] + w[-2:]
print(first_last_two("bharat"))
print(first_last_two("python"))
```

```python
# 24 

def insert_middle(original, insert):
    mid = len(original) // 2
    return original[:mid] + insert + original[mid:]

print(insert_middle("python", "123"))
print(insert_middle("bharat", "shyam"))
```
