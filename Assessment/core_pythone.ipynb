Healthcare industry

```python
class clinicappointment:
    def __init__(self):
        self.slots = ["10am", "11am", "12pm", "2pm", "3pm"]
        self.doctors = ["dr.vyas", "dr.patel", "dr.gandhi"]
        self.appointments = {doc: {slot: [] for slot in self.slots} for doc in self.doctors}

    def book_appointment(self):
        name = input("enter your name: ")
        age = int(input("enter your age: "))
        mobile = input("enter your mo_number: ")

        print("name",name,
              "age",age,
              "mobile",mobile)
    

        print("\ndoctors:")

        for i, d in enumerate(self.doctors, 1):
            print(i, d)
        choice_doctor = self.doctors[int(input("choice doctor(1-3)")) - 1]

        print("\nchoice slot")
        choice_slot = self.slots[int(input("choice slots(1-5)")) - 1]

        if len(self.appointments[choice_doctor][choice_slot]) < 3:
            self.appointments[choice_doctor][choice_slot].append({"name": name, "age":age, "mobile":mobile})
            print("appointment booked with", choice_doctor, "at", choice_slot)
        else:
            print("slot full try another.")


    def view(self):
        mobile = input("enter your currect mo_number: ")
        for d in self.doctors:
            for s in self.slots:
                for p in self.appointments[d][s]:
                    if p["mobile"] == mobile:
                        print(p["name"],"-", d, "-", s)
                        return
                    else:
                        print("no appointment found:")



    def cancel(self):
        mobile = input("enter your currect mo.number: ")
        for d in self.doctors:
            for s in self.slots:
                for p in self.appointments[d][s]:
                    if p["mobile"] == mobile:
                        self.appointments[d][s].remove(p)
                        print("appointment cancelled.")
                        return
                    else:
                        print("no appointment found.")

    def availability(self):
        for d in self.doctors:
            print("\n", d)
            for s in self.slots:
                left = 3 - len(self.appointments[d][s])
                print(" ", s, "-", left, "slot left")

c = clinicappointment()

while True:
        print("\n 1.book, 2.view, 3.cancel, 4.Availability  5.exit")
        choice = input("enter choice: ")
        if choice == "1":
            c.book_appointment()
        elif choice == "2":
            c.view()
        elif choice == "3":
            c.cancel()
        elif choice == "4":
            c.availability()
        elif choice == "5":
            break
        else:
            print("invalid choicess.")            
                                                       
```

School managment System

```python
class Schoolmanagement:
    def __init__(self):
        self.students = {}
        self.student_id = 1

    def admission(self):
        name = input("enter your name")
        age = int(input("enter your age: "))
        std_class = int(input("enter your class: "))
        mobile = input("enter mobile num: ")

        if  age < 5 or age > 18:
            print("limit of  age bitween 5 - 18 ")
            return
        if len(mobile) != 10 or not mobile.isdigit():
            print("mobile number aloud 10 digit")
            return

        
        self.students[self.student_id] = {"name": name, "age": age, "class": std_class, "mobile": mobile}
        print(f"Admission successfull student_id : {self.student_id}")
        self.student_id += 1


    def view(self):
        s_id = int(input(" enter student id: "))
        if s_id in self.students:
            s = self.students[s_id]
            print(f"\n  student details (id {s_id}): name: {s['name']}, age: {s['age']}, class: {s['class']},mobile: {s["mobile"]}")
        else:
            print("not get student")

    def update(self):
        s_id = int(input("enter student id:"))
        if s_id in self.students:
            print("1. mobile update  \n2. update class")
            choice = input("enter choice 1 ya 2: ")
            if choice == "1":
                mob = input("enter currect mobile num: ")
                if len(mob) == 10 and mob.isdigit():
                    self.students[s_id]['mobile'] = mob
                    print("succes mobile update.")
                else:
                    print("invalid mobile.")

            elif choice == "2":
                cls = int(input("enter nwe class(1-12): "))
                if 1 <= cls <= 12:
                    self.students[s_id]['class'] = cls
                    print("succes class update.")
                else:
                    print("invalid class")
            else:
                print("not found.")


    def remove_std(self):
        s_id = int(input("enter currect student id: "))
        if s_id in self.students:
            del self.student_id[s_id]
            print("removed student.")
        else:
            print("not found.")

    def run(self):
        while True:
            print("\n 1.new Admission \n2.view student \n3.update \n4.remove s \n5. exit")
            choice = input("your choice (1-5): ")
            if choice == "1":
                self.admission()
            elif choice == "2":
                self.view()
            elif choice == "3":
                self.update()
            elif choice == "4":
                self.remove_std()
            elif choice == "5":
                print("exit")
                break
            else:
                print("wrong choice.")

s = Schoolmanagement()
s.run()                                                                    
```

Transport Reservation System(bus ticketing)

```python
class BusBooking:
    def __init__(self):
        self.routes = {
            "Mumbai to Pune" : 500,
            "Delhi to Jaipur" : 600,
            "Ahmedabad to Mumbai" : 400,
            "Ahmedabad to Baroda" : 200
        }
        self.booked_tickets = []
        self.booking_id = 1001
        self.total_seats = 60
        self.seats = {seat : None for seat in range(1, self.total_seats + 1)}


    def show_routes(self):
        print("Pleaes find below available routs: ")
        for route, price in self.routes.items():
            print(f"{route} - {price}")

    def show_available_seats(self):
        available_seats = [seat for seat, status in self.seats.items() if status is None]
        print(f"available seats are {available_seats}")
        return available_seats
    

    def ticket_booking_process(self):
        print("Please Enter Below details for Booking : ")
        
        Passanger_name = input("Enter your Name: ")
        if not Passanger_name:
            print("filed should not be blank")
            return
        Passanger_age = int(input("Please enter your age: "))
        if Passanger_age < 0:
            print("Pleaes Enter Valid Age")
            return
        Passanger_Mobile = input("Enter your Phone Number : ").strip()
        if not (Passanger_Mobile.isdigit()) and len(Passanger_Mobile) == 10:
            print("Enter Valid Mobiel Number")
            return
        Booking_number = self.booking_id
        self.booking_id += 1


        self.show_routes()

        route_selection = input("Enter the route you want to travel: ").strip()
        if route_selection not in self.routes:
            print("Entered Route is not available : ")
            return
        price = self.routes[route_selection]

        available_seats = self.show_available_seats()

        seat_choice = int(input("Entre Seat number from the available seats: "))
        if seat_choice not in available_seats:
            print(f"Seat Number {seat_choice} has been already boooked")


        confirmation = {
            "booking_id" : Booking_number,
            "name" : Passanger_name,
            "age" : Passanger_age,
            "mobile" : Passanger_Mobile,
            "route" : route_selection,
            "price" : price,
            "seat" : seat_choice
        }

        self.booked_tickets.append(confirmation)

        print(f"booking has been confirmed for {Passanger_name} with Booking Number {Booking_number} and seat number {seat_choice}")

    def view_booking_with_booking_id(self):
        entered_id = int(input("Enter your Booking Number : "))
        for details in self.booked_tickets:
            if details ["booking_id"] == entered_id:
                print("\n Booking Details found: ")
                print(f"Passanger Name : {details['name']}")
                return
            
    def cancle_booking_with_booking_Id(self):
        entered_id_c = int(input("Enter your Booking Number : "))
        for i, details in enumerate(self.booked_tickets):
            if details ["booking_id"] == entered_id_c:
                Change_seat_status = details["seat"]   # Change sear status to available
                self.seats[Change_seat_status] = None  # seat stauts cahnged to none

                self.booked_tickets.pop(i)   # to remove booking

                print(f"Booking has been cancelled for {entered_id_c}")


Travell = BusBooking()
Travell.ticket_booking_process()
Travell.view_booking_with_booking_id()
Travell.cancle_booking_with_booking_Id()
```
